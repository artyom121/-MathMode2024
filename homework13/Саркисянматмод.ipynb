```python
##задача 2 Автобусы некоторого маршрута идут строго по расписанию и интервалом 5 минут. Составьте функцию плотности случайной величины X – времени ожидании очередного автобуса пассажиром, который наугад подошёл к остановке. Найдите вероятность того, что он будет ждать автобус не более 2 минут.
```

```python
import numpy as np

def wait_time_pdf(x):
    a, b = 0, 5  # Границы интервала
    if a <= x <= b:
        return 1 / (b - a)
    else:
        return 0

# Вычисление вероятности P(X <= 2)
prob = np.vectorize(wait_time_pdf)(np.linspace(0, 2, 1000)).mean() * 2

print(f"Вероятность ждать автобус не более 2 минут: {prob:.4f}")
```

```python
##Задача 3 Рост мальчиков возрастной группы 15 лет есть нормально распределённая случайная величина X с математическим ожиданием роста 161 см и среднеквадратичным отклонением 4 см. Какую долю костюмов для мальчиков, имеющих рост от 152 до 158 см, нужно предусмотреть в объёме производства для данной возрастной группы?
```

```python
import scipy.stats as stats

mu = 161  
sigma = 4  

lower_bound = 152
upper_bound = 158

z_lower = (lower_bound - mu) / sigma
z_upper = (upper_bound - mu) / sigma

probability = stats.norm.cdf(z_upper) - stats.norm.cdf(z_lower)
print(f"Доля костюмов для мальчиков с ростом от 152 до 158 см: {probability:.4f}")
```

```python
##задача 4 На машиностроительном заводе в процессе производства за неделю обнаруживается в среднем 100 бракованных деталей со среднеквадричным отклонением в 12 деталей. Рассчитайте следующее: 1) вероятность того, что за неделю будет 20-30 бракованных деталей, 2) вероятность того, что за день будет хотя бы одна забракованная деталь.
```

```python
import scipy.stats as stats

mu = 100 
sigma = 12 

lower_bound = 20
upper_bound = 30

z_lower = (lower_bound - mu) / sigma
z_upper = (upper_bound - mu) / sigma

probability_20_to_30 = stats.norm.cdf(z_upper) - stats.norm.cdf(z_lower)
print(f"Вероятность того, что за неделю будет 20-30 бракованных деталей: {probability_20_to_30:.4f}")

mu_daily = mu / 7  
p_at_least_one_defective = 1 - stats.norm.cdf(0, mu_daily, sigma)
print(f"Вероятность того, что за день будет хотя бы одна забракованная деталь: {p_at_least_one_defective:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Параметры нормального распределения
mean = 0
std_deviation = 4

# Создание нормального распределения
normal_dist = stats.norm(loc=mean, scale=std_deviation)

# Генерация значений для построения графиков
x = np.linspace(-20, 20, 1000)
pdf = normal_dist.pdf(x)
cdf = normal_dist.cdf(x)

# Построение графиков
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, color='blue')
plt.title('PDF нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Плотность вероятности')

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='red')
plt.title('CDF нормального распределения')
plt.xlabel('Значение')
plt.ylabel('Вероятность')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Границы интервала времени между автобусами
a = 0
b = 5

# Функция плотности вероятности (PDF) для равномерного распределения
def uniform_pdf(x, a, b):
    return 1/(b-a) if a <= x <= b else 0

# Функция функции распределения вероятностей (CDF) для равномерного распределения
def uniform_cdf(x, a, b):
    if x < a:
        return 0
    elif a <= x <= b:
        return (x - a)/(b - a)
    else:
        return 1

# Создание значений для построения графиков
x = np.linspace(-1, 6, 1000)
pdf = [uniform_pdf(val, a, b) for val in x]
cdf = [uniform_cdf(val, a, b) for val in x]

# Построение графиков
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf, color='blue')
plt.title('PDF времени ожидания автобуса')
plt.xlabel('Время ожидания (мин)')
plt.ylabel('Плотность вероятности')

plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='red')
plt.title('CDF времени ожидания автобуса')
plt.xlabel('Время ожидания (мин)')
plt.ylabel('Вероятность')

plt.tight_layout()
plt.show()
```
