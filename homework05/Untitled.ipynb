```python
from PIL import Image 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = Image.open('rudn.jpg') 
x = np.array(img, dtype=np.float32) 
y = img.convert('L')
x = np.array(list(y.getdata(band=0)), float)
x.shape= (y.size[1], y.size[0])
x = np.matrix(x)

plt.imshow(x, cmap='gray');
```

```python
x.shape
U, S, V = np.linalg.svd(x, full_matrices=False)
k = [1, 2, 10, 30, 50, 100]
for i in k:
    rudn = np.matrix(U[:, :i]) * np.diag(S[:i]) * np.matrix(V[:i, :])
    plt.imshow(rudn, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()

Image.fromarray(np.asarray(rudn, dtype=np.uint8)).save('rudn.png')
```
