№6 задание
Создайте функцию, которая создает  n×n  матрицу с  (i,j) -элементом, равным  i+j .

```python
import numpy as np

def create_matrix(n):
    matrix = np.fromfunction(lambda i, j: i + j, (n, n), dtype=int)
    return matrix

n = 4
result = create_matrix(n)
print(result)
```

№2 задание
Создайте semilogy график (с логарифмической шкалой для оси y и линейной шкалой для оси x) относительной разницы функций  1/(1+x^2)  и  1/x^2  на интервале  [5,25] . (Относительная разница чисел  a  и  b  определяется как  |1−a/b| )

```python
import numpy as np
import matplotlib.pyplot as plt

x_min = 5
x_max = 25

x = np.linspace(x_min, x_max, 100)

f1 = 1 / (1 + x**2)
f2 = 1 / x**2

relative_diff = np.abs(1 - f1/f2)

plt.semilogy(x, relative_diff)
plt.xlabel('x')
plt.ylabel('Относительная разница')
plt.title('Относительная разница функций 1/(1+x^2) и 1/x^2')
plt.grid(True)
plt.show()
```

№4 задание
Создайте цветовой график  sin(x)sin(y)  на интервале  [−π,π]×[−π,π] .

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)

X, Y = np.meshgrid(x, y)

Z = np.sin(X) * np.sin(Y)

plt.pcolormesh(X, Y, Z, cmap='jet')
plt.colorbar(label='Значения функции')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График sin(x)sin(y)')
plt.show()
```

№2 задание
Создайте случайный массив  3×5  с помощью функции np.random.rand(3, 5) и вычислите: сумму всех элементов, сумму строк и сумму столбцов. (Подобно тому, как sorted имеет необязательный аргумент key=, многие функции Numpy имеют необязательный аргумент axis=)

```python
import numpy as np

array = np.random.rand(3, 5)
print("Случайный массив:")
print(array)

total_sum = np.sum(array)
print("\nСумма всех элементов: ", total_sum)

row_sums = np.sum(array, axis=1)
print("\nСуммы строк: ", row_sums)

column_sums = np.sum(array, axis=0)
print("\nСуммы столбцов: ", column_sums)
```

Создайте случайный массив  5×5 , используя функцию np.random.rand(5, 5). Отсортируйте строки по второму столбцу. Используйте для этого slicing + argsort + indexing.

```python
import numpy as np

array = np.random.rand(5, 5)
print("Исходный массив:")
print(array)

sorted_array = array[array[:, 1].argsort()]
print("\nМассив после сортировки по второму столбцу:")
print(sorted_array)
```

№4 задание
Посчитайте две матрицы:  B=ATA  и  C=AAT . Что вы можете о них сказать? Какого они размера? В чём их сходство и отличие? Найдите сумму диагональных элементов матриц  B  и  C . Посчитайте детерминант и ранг матриц  A ,  B  и  C , посмотрите на спектр (набор собственных значений) матриц  B  и  C , какие выводы вы можете сделать? (подсказка: используйте готовые функции из библиотеки np.linalg).

```python
import numpy as np

A = np.random.rand(3, 3)
print("Матрица A:")
print(A)

B = np.dot(A.T, A)
print("\nМатрица B:")
print(B)

C = np.dot(A, A.T)
print("\nМатрица C:")
print(C)

print("\nРазмер матрицы B:", B.shape)
print("Размер матрицы C:", C.shape)

print("\nB и C являются квадратными матрицами.")
print("Матрица B получается умножением транспонированной матрицы A на саму матрицу A.")
print("Матрица C получается умножением матрицы A на транспонированную матрицу A.")

print("\nСумма диагональных элементов матрицы B:", np.trace(B))
print("Сумма диагональных элементов матрицы C:", np.trace(C))

print("\nДетерминант матрицы A:", np.linalg.det(A))
print("Детерминант матрицы B:", np.linalg.det(B))
print("Детерминант матрицы C:", np.linalg.det(C))
print("Ранг матрицы A:", np.linalg.matrix_rank(A))
print("Ранг матрицы B:", np.linalg.matrix_rank(B))
print("Ранг матрицы C:", np.linalg.matrix_rank(C))

eigenvalues_B = np.linalg.eigvals(B)
eigenvalues_C = np.linalg.eigvals(C)

print("\nСпектр матрицы B:")
print(eigenvalues_B)

print("\nСпектр матрицы C:")
print(eigenvalues_C)
```
