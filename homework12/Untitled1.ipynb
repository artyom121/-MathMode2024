задание 1 Создайте массив нулей размерностью 8х8х3. Создайте массив, заполненный значениями 150, размерностью 8х8х3. Создайте массив, заполненный значениями 250, размерностью 8х8х3. Выведите на экран все 3 массива в виде изображений при помощи библиотеки matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

arr_zeros = np.zeros((8, 8, 3), dtype=np.uint8)

arr_150 = np.full((8, 8, 3), 150, dtype=np.uint8)

arr_250 = np.full((8, 8, 3), 250, dtype=np.uint8)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(arr_zeros)
plt.title("Array of Zeros")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(arr_150)
plt.title("Array of 150")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(arr_250)
plt.title("Array of 250")
plt.axis('off')

plt.show()
```

Задача 2
При помощи библиотек numpy и matplotlib создайте и выведите на экран 5 изображений размером 100х100 пикселей белого, черного, красного, синего и зеленого цветов.


```python
import numpy as np
import matplotlib.pyplot as plt

white_image = np.full((100, 100, 3), 255, dtype=np.uint8)
black_image = np.zeros((100, 100, 3), dtype=np.uint8)
red_image = np.zeros((100, 100, 3), dtype=np.uint8)
red_image[:, :, 0] = 255
blue_image = np.zeros((100, 100, 3), dtype=np.uint8)
blue_image[:, :, 2] = 255
green_image = np.zeros((100, 100, 3), dtype=np.uint8)
green_image[:, :, 1] = 255

plt.figure(figsize=(15, 5))

plt.subplot(1, 5, 1)
plt.imshow(white_image)
plt.title("White Image")
plt.axis('off')

plt.subplot(1, 5, 2)
plt.imshow(black_image)
plt.title("Black Image")
plt.axis('off')

plt.subplot(1, 5, 3)
plt.imshow(red_image)
plt.title("Red Image")
plt.axis('off')

plt.subplot(1, 5, 4)
plt.imshow(blue_image)
plt.title("Blue Image")
plt.axis('off')

plt.subplot(1, 5, 5)
plt.imshow(green_image)
plt.title("Green Image")
plt.axis('off')

plt.show()
```
