```python
Задача 2

Используя библиотеку OSMnx проложите кратчайший по расстоянию маршрут от Вашего дома до работы (или учебного заведения) и выведите его на экран. 
Тип сети network_type используйте “drive”, т.е. для автомобилей.
```

```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

# Адрес дома
home_address = "Тверская улица, 6с2, Москва, Россия"

# Адрес работы
work_address = "Кутузовский проспект, 36, Москва, Россия"

# Получаем координаты дома и работы из адресов
home_coords = ox.geocode(home_address)
work_coords = ox.geocode(work_address)

# Разделяем координаты на широту и долготу
home_lat, home_lon = home_coords
work_lat, work_lon = work_coords

# Загружаем данные о дорожной сети из OpenStreetMap
graph = ox.graph_from_point(home_coords, dist=10000, network_type='drive')

# Находим ближайшие узлы к дому и работе
origin_node = ox.nearest_nodes(graph, home_lon, home_lat)
destination_node = ox.nearest_nodes(graph, work_lon, work_lat)

# Находим кратчайший путь между домом и работой
shortest_route = nx.shortest_path(graph, origin_node, destination_node, weight='length')

# Визуализируем кратчайший маршрут на карте
fig, ax = ox.plot_graph_route(graph, shortest_route, route_linewidth=6, node_size=0)

# Визуализируем начальную и конечную точки
origin_point = (graph.nodes[origin_node]['y'], graph.nodes[origin_node]['x'])
destination_point = (graph.nodes[destination_node]['y'], graph.nodes[destination_node]['x'])
nc = ox.plot.get_node_colors_by_attr(graph, 'default', cmap='viridis')
ox.plot_graph(graph, ax=ax, node_color=nc, node_edgecolor='k', node_zorder=3, nodelist=[origin_node, destination_node], node_size=100)

plt.show()








```
